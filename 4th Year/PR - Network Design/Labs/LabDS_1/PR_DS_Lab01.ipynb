## Vectorizari

In contextul limbajelor high-level (precum Python, MATLAB, R), vectorizarea descrie utilizarea codului optimizat si pre-compilat pentru a realiza operatii asupra unor secvente de date. Acest lucru presupune executia in-place a operatiei asupra tuturor elementelor colectiei, fara a mai fi nevoie de o iteratie explicita.

Putem observa eficienta vectorizarii prin aplicarea functiei sum() in urmatorul exemplu:

```python
import timeit

rand_arr = np.random.rand(30, 30)
rand_arr_vect = rand_arr
rand_arr_iter = rand_arr

def add_vect(arr):
  sum = np.sum(arr)

def add_iter(arr):
  sum = 0
  for num in arr:
    sum += num
```

```python
%timeit add_vect(rand_arr_vect)
```

```python
%timeit add_iter(rand_arr_iter)
```

#### Cum manipulam un DataFrame folosind functionale?

Acest honey farm colecteaza valorile bandwidth-ului unui atac in bps (biti per secunda), ceea ce le face greu de urmarit.

Vom crea o functie care va converti bandwidth-ul la Gbps (gigabiti per secunda) si o vom aplica pe datele din dataframe folosind *apply()*.

```python
honeypot = pd.read_csv("AWS_Honeypot.csv")

def convert_to_gbps(bandwidth):
  return round((float) (bandwidth / (10 ** 9)), 2)

honeypot["src"] = honeypot["src"].apply(lambda x: convert_to_gbps(x))
honeypot.head()
```

#### Cum intelegem datele la nivel de grup? *(groupby)*

Dorim sa interpretam datele in functie de protocolul folosit in atac:

```python
for group, frame in honeypot.groupby("proto"):
  print(str(group) + ": " + str(len(frame)) + " records")
```

##### **GroupBy** este un proces ce implica cel putin unul din pasii *split-apply-combine*:

*   *split*: datele se impart in grupuri dupa un criteriu
*   *apply*: se aplica o functie pe fiecare grup
*   *combine*: rezultatele sunt agregate intr-o structura

La pasul apply, putem folosi groupby() impreuna cu una din functiile:

*Aggregate*

Pasam un dictionar avand numele coloanelor de interpretat drept chei si functiile necesare interpretarii drept valori.

Analizam in continuare grupurile, calculand bandwidth-ul mediu si latitudinea minima pentru fiecare:

```python
honeypot.groupby("proto").agg({"src": np.nanmean, "latitude": np.min})
```

*Filter*

Pasam o functie care se aplica pe fiecare grup si intoarce True sau False, in functie de rezultatul grupului la filtru.

Vrem sa gasim informatii doar despre atacurile din grupurile care au media bandwidth-ului mai mare de 2.5 Gbps:

```python
# gasim informatii doar despre atacurile din grupurile care au media
# bandwidth-ului mai mare de 2.5 Gbps
bw = honeypot.groupby("proto").filter(lambda df: np.nanmean(df["src"]) > 2.5)
bw.reset_index(inplace=True)
bw.drop(columns="index")
```

*Apply*

Pasam o functie care se va aplica pe fiecare grup, oferindu-ne posibilitatea de a pastra rezultatele obtinute in cadrul aceluiasi dataframe.

Intentionam sa grupam atacurile dupa protocolul folosit si sa aflam care este deviatia standard a fiecarui atac fata de media grupului din care face parte.

```python
df = honeypot[['src', 'proto']]

def compute_std_dev(group):
  avg = np.nanmean(group['src'])
  group['avg'] = avg
  group['std_dev'] = np.abs(avg - group['src'])
  return group

df.groupby('proto').apply(compute_std_dev)
```

## Valori Lipsa

In practica, putem intalni seturi de date cu valori lipsa. Un data scientist va deduce semnificatia acestora sau le va elimina in procesul de *data cleaning*.

NumPy foloseste ***np.nan*** pentru a indica valori lipsa. NaN (Not a Number) este o valoare de tip float in standardul IEEE, fapt ce permite utilizarea operatiilor matematice asupra structurilor care o contin. Toate operatiile efectuate asupra unor structuri ce contin NaN vor intoarce, la randul lor, NaN.

*Atentie!* np.nan este diferit de None, obiect de tip NoneType care nu accepta operatii matematice.

```python
a = None
b = np.nan
print(a is b)
```

```python
floats = np.array([2, np.nan, -1.3])
print(np.sum(floats))
```

Pentru a verifica daca o structura contine sau nu valori lipsa, putem folosi *isnull()*.



### 7. Calculati corelatia dintre rating-ul de pe IMDB si numarul de premii castigate de un film, alaturi de nivelul de incredere aferent.

#### *Coeficientul de Corelatie Pearson - r*

Coeficientul de corelatie Pearson masoara dependenta liniara a doua seturi de date X si Y.

Valoarea acestuia variaza intre -1 si 1, valoarea de 0 reprezentand faptul ca nu exista nicio corelatie intre cele doua seturi de date.

O valoare pozitiva indica faptul ca daca valorile din setul de date X cresc, atunci valorile din setul de date Y cresc, avand o relatie direct proportionala.

O valoare negativa indica faptul ca daca valorile din setul de date Y cresc, atunci valorile din setul de date Y scad, avand o relatie invers proportionala.

Coeficientul de corelatie Pearson poate reflecta numai o corelatie liniara intre doua seturi de date, neluand in considerare alte tipuri de corelatii intre variabile.

Exemplu: Ne asteptam ca varsta si inaltimea unui grup de adolescenti sa aiba un coeficient de corelare Pearson mai mare decat 0, dar mai mic decat 1 (din cauza ca 1 ar insemna o corelatie perfecta ce este imposibila in acest caz).

#### *Nivelul de incredere  - p-value*

Nivelul de incredere - p-value - releva probabilitatea ca doua seturi de date ce nu au niciun tip de corelatie intre ele sa produca un coeficient de corelatie Pearson egal cu cel calculat intre cele doua.

##### **Atentie!** :  p-value se exprima cu o valoare situata intre 0 si 1, nu folosind procente, deoarece reprezinta in esenta o probabilitate.

Hint: scipy.stats.pearsonr(x, y)



```python
!pip install scipy
import scipy
from scipy.stats import pearsonr

all_festivals = pd.concat([golden_globes, bafta])
ratings = []
awards = []

for _, movie in movies.iterrows():
    ratings.append(movie['avg_vote'])
    lst = all_festivals[all_festivals['original_title'] == movie['original_title']]
    crt_no_awards = 0
    for _, festival in lst.iterrows():
        crt_no_awards += festival['Winner']
    awards.append(crt_no_awards)

p_value = scipy.stats.pearsonr(ratings, awards)
print(f'P-Value between IMDB rating and total number of awards is {p_value}.')
```

### 8. Calculati deviatia standard a rating-ului fiecarui film fata de nota medie a filmelor din acelasi gen.



Rezultatul poate fi fie o noua coloana, fie un Series.

```python
genre_groups = movies['genre']

genres = []
for genre_group in genre_groups:
    crt_group = genre_group.split(',')
    for genre in crt_group:
        genre = genre.strip()
        if genre not in genres:
            genres.append(genre)

means = {}
for genre in genres:
    count = 0
    crt_sum = 0
    for _, movie in movies.iterrows():
        if genre in movie['genre']:
            count += 1
            crt_sum += movie['avg_vote']
    crt_genre_mean = crt_sum / count
    means[genre] = crt_genre_mean

deviations = {}
for _, movie in movies.iterrows():
    crt_movie_genres = movie['genre'].split(',')
    crt_deviations = []
    for crt_genre in crt_movie_genres:
        crt_genre = crt_genre.strip()
        crt_dev = abs(means[crt_genre] - movie['avg_vote'])
        crt_deviations.append((crt_genre, crt_dev))
    deviations[movie['original_title']] = crt_deviations

print(deviations)
```

### 9. Filtrati dataframe-ul movies, astfel incat sa contina doar filmele cu media rating-ului genului mai mare decat 8.65.

```python
query_movies2 = []
for genre in genres:
    if means[genre] > 8.65:
        for _, movie in movies.iterrows():
            crt_genres = movie['genre'].split(',')
            for crt_genre in crt_genres:
                crt_genre = crt_genre.strip()
                if genre == crt_genre:
                    query_movies2.append(movie)
                    break

query_movies2_df = pd.DataFrame(data=query_movies2, columns=movies.columns)
query_movies2_df.head()
```

### 10. Extrageti deceniul cu cea mai mare medie a rating-ului filmelor.

*Hint: apply, groupby*

```python
decades = {}
for _, movie in movies.iterrows():
    crt_decade = movie['year'] // 10 * 10
    if crt_decade not in decades:
        decades[crt_decade] = (movie['avg_vote'], 1)
    else:
        decades[crt_decade] = (decades[crt_decade][0] + movie['avg_vote'], decades[crt_decade][1] + 1)

for k, v in decades.items():
    decades[k] = v[0] / v[1]

max_idx = max(decades, key=decades.get)
print(f'Decade with highest average ratings is {max_idx}.')
```

### 11. Care pereche actor-regizor are cele mai multe colaborari?

```python
colaborations = {}

for _, movie in movies.iterrows():
    crt_actors = movie['actors'].split(',')
    for actor in crt_actors:
        actor = actor.strip()
        name = str(movie['director']) + '_' + str(actor)
        if name not in colaborations:
            colaborations[name] = 1
        else:
            colaborations[name] = colaborations[name] + 1

max_idx = max(colaborations, key=colaborations.get).split('_')
print(f'Most frequent colaboration is between director {max_idx[0]} and actor {max_idx[1]}.')
```
